#### Data Cleaning 


we replace the null value in age column instead deleting it with the mean age with respect to passenger class


```python
sns.boxplot(x='Pclass', y='Age',data=df,palette='rainbow')
```

```python
#  function to replace the null values 
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df['Age']= df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
# checking heat map 
sns.heatmap(df.isnull(), yticklabels= False)
```

```python
df.drop('Cabin',axis=1,inplace=True)
```
